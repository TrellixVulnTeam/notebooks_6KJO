### get state level pop  estimates

```python
state_fips = pd.read_csv("https://gist.githubusercontent.com/dantonnoriega/bf1acd2290e15b91e6710b6fd3be0a53/raw/11d15233327c8080c9646c7e1f23052659db251d/us-state-ansi-fips.csv")
```

```python
state_fips["code"]=state_fips[' st'].astype(str)
```

```python
long_fips_states = state_fips[state_fips['code'].str.len()==2]
short_fips_states = state_fips[state_fips['code'].str.len()==1]
short_fips_states["code"]="0"+short_fips_states["code"]
```

```python
all_state_fips = short_fips_states.append(long_fips_states)
```

```python
pop_var = 'B01001_001E'
state_codes=all_state_fips["code"]
```

```python
state_pop_frames = []

for state in state_codes:
    state_pop_frames.append(
    pd.read_json(
    "https://api.census.gov/data/2020/acs/acs5?get=NAME," 
        + pop_var 
        + "&for=state:"
        + str(state)
        +"&key=aa5d6d0fefa3172a9f363636440111d6584a0ac4").assign(state=state))
```

```python
states_pop=pd.concat(state_pop_frames)
```

```python
states_pop.columns=['state','value','fips','fips2']
```

```python
states_pop_clean=states_pop[states_pop['state']!="NAME"]
states_pop_clean.drop(['fips2'],axis=1).rename(columns={"value": "pop"})
```

```python
oews_raw=pd.read_csv("data/raw/states_rnwages.csv",thousands=',')
```

```python
oews_raw.columns=oews_raw.columns.str.lower()
```

```python
rns_emp=oews_raw[['area_title','area','prim_state','tot_emp']]

rns_emp['fips']=rns_emp['area'].astype(str)
```

```python
rns_emp.drop(['area'],axis=1)
shortfips_rns=rns_emp[rns_emp['fips'].str.len()==1]
shortfips_rns["fips"]="0"+shortfips_rns["fips"]
longfips_rns=rns_emp[rns_emp['fips'].str.len()==2]
rns_final = shortfips_rns.append(longfips_rns)
```

```python
rns_by_pop=pd.merge(rns_final,states_pop_clean,on="fips",how="left").drop(['area','fips2','state'],axis=1).rename(columns={"value": "pop"})
```

```python
rns_bypop_clean=rns_by_pop[rns_by_pop['pop'].notnull()]
```

```python
rns_bypop_clean['pop']=rns_bypop_clean['pop'].astype(int)
```

```python
rns_bypop_clean['nurses_per1k']=1000*rns_bypop_clean['tot_emp']/rns_bypop_clean['pop']
```
