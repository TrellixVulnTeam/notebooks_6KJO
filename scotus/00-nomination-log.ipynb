```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
pd.options.display.max_columns = 100
pd.options.display.max_rows = 1000
pd.options.display.max_colwidth = None
```

```python
url = 'https://www.senate.gov/legislative/nominations/SupremeCourtNominations1789present.htm'
```

```python
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')
dfs = pd.read_html(page.text)
```

```python
src = pd.DataFrame(dfs[2])
```

```python
src.drop([0], inplace=True)
src.drop([1, 3, 5, 7, 9], axis=1, inplace=True)
```

```python
src.columns = ['nominee', 'replacing', 'nomination_date', 'vote', 'result', 'result_date']
```

```python
df = src[~src['nominee'].str.contains("President")].copy()
```

```python
# df['result_date'] = df['result_date'].fillna('')
# df['vote'] = df['vote'].fillna('')
# df['result'] = df['result'].fillna('')
```

```python
df[['vote_result', 'vote_number']] = df['vote'].str.split(" ", n=1, expand=True)
```

```python
df['vote_number'] = df['vote_number'].str.replace('None', "", regex=False)
```

```python
df[['yes', 'no']] = df['vote_result'].str.split('-', expand=True)
```

```python
df['nominee'] = df['nominee'].str.replace('\d+', '', regex=True)
```

```python
df['yes'] = df['yes'].str.replace('V', '100', regex=False)
df['no'] = df['no'].fillna(0)
```

```python
df['nomination_date'] = pd.to_datetime(df['nomination_date'])
df['result_date'] = pd.to_datetime(df['result_date'])
```

```python
df['nomination_year'] = df['nomination_date'].dt.year
df['result_year'] = df['result_date'].dt.year
```

```python
got_vote = df[~df['vote_result'].isnull()].copy()
```

```python
got_vote['result_year'] = got_vote['result_year'].astype(str).str.replace('.0', '', regex=False)
```

```python
got_vote.drop(['vote', 'vote_result', 'vote_number'], axis=1, inplace=True)
```
