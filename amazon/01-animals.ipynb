### Get a dataframe listing all the words mentioned in the 'title' column and clean it up

```python
words = (
    pd.DataFrame(
        df["title"]
        .str.replace(",", "", regex=False)
        .str.replace("|", "", regex=False)
        .str.replace("&", "", regex=False)
        .str.replace("-", "", regex=False)
        .str.strip()
        .str.split(expand=True)
        .stack()
        .value_counts(),
        columns=["count"],
    )
    .reset_index()
    .rename(columns={"index": "word"})
)
```

### Lowercase words

```python
words["word"] = words["word"].str.lower()
```

### Make the list of animals lowercase, too

```python
for i in range(len(animals)):
    animals[i] = animals[i].lower()
```

### Which product words match the animals list? 

```python
words_animal = words[words["word"].isin(animals)]
```

### Group by animal and sum

```python
words_animal = (
    words_animal.groupby("word")
    .agg({"count": sum})
    .reset_index()
    .sort_values("count", ascending=False)
)
```

### Share for each animal out of all the animals

```python
words_animal["share"] = (
    (words_animal["count"] / words_animal["count"].sum()) * 100
).round(2)
```
