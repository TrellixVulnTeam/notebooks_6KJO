## Grid story feed

```python
import requests
import pandas as pd
```

```python
params = {
    'query': '{"authorSlug":"matt-stiles","feature":"results-list","feedOffset":0,"feedSize":100}',
    'filter': '{content_elements{_id,credits{by{_id,additional_properties{original{role}},name,slug,type,url}},description{basic},display_date,first_publish_date,headlines{basic,mobile},label{basic{display,text,url}},promo_items{basic{resized_params{158x89,274x154},type,url},lead_art{streams{stream_type,url},subtype,type}},revision{published},taxonomy{primary_section{_id,name}},type,websites{the-summit{website_url}}}}',
    'd': '86',
    '_website': 'the-summit',
}

response = requests.get('https://www.grid.news/pf/api/v3/content/fetch/story-feed-author', params=params)
```

```python
src = pd.DataFrame(response.json()['content_elements'])
```

```python
src.head()
```

```python
src['description'] = pd.json_normalize(src['description'])
```

```python
src['headline'] = pd.json_normalize(src['headlines'])['basic']
```
