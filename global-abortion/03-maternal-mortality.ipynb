```python
conditions = [
    (abortion_countries['category'] == '1348'),
    (abortion_countries['category'] == '1349'),
    (abortion_countries['category'] == '1350'),
    (abortion_countries['category'] == '1351'),
    (abortion_countries['category'] == '1352')
]

values = ['Prohibited broadly','Permitted to protect life of pregnant person','Permitted to protect health','Permitted under broad socioeconomic grounds','Permitted with varying gestational limits']

abortion_countries['abortion_rights_text'] = np.select(conditions, values)

abortion_countries.head()
```
