### Grab the latest poll trend from 538

```python
trend_url = (
    "https://projects.fivethirtyeight.com/biden-approval-data/approval_topline.csv"
)
```

```python
trend_src = pd.read_csv(trend_url)
```

```python
trend_src.tail()
```

```python
trend_src.subgroup.value_counts()
```

```python
trend_src.columns
```

```python
trend_src.to_csv("data/processed/five_thirty_eight_approval_topline.csv", index=False)
```

```python
for g in trend_src["subgroup"].unique():
    trend_src[trend_src["subgroup"] == g].to_csv(
        f"data/processed/five_thirty_eight_approval_topline_{g.replace(' ', '_').lower()}.csv",
        index=False,
    )
```
