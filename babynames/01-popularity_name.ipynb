```python
df = pd.read_csv("data/processed/names_births.csv", dtype={"year": str})
```

```python
df[df["year"] == "2020"].sort_values("fprop", ascending=False)
```

```python
names = ["Briar"]
```

```python
name = df[(df["name"].isin(names)) & (df["sex"] == "F")].reset_index(drop=True)
```

```python
name.sort_values("year", ascending=False).head()
```

```python
line_chart = (
    alt.Chart(name)
    .mark_line(color="pink", size=4)
    .encode(
        x=alt.X(
            "year:T",
            title="Year",
            axis=alt.Axis(tickCount=10, grid=False),
        ),
        y=alt.Y(
            "fprop",
            title="Percentage of all names",
            stack=None,
            axis=alt.Axis(
                tickSize=0,
                domainOpacity=0,
                tickCount=6,
                offset=4,
                gridWidth=0.6,
                gridColor="#dddddd",
                format="%",
            ),
        ),
        color=alt.Color("name", title=" "),
    )
)


chart = (
    (line_chart)
    .properties(
        width=1000,
        height=500,
        title="Popularity of the name Briar",
    )
    .configure_view(strokeOpacity=0)
    .configure_legend(orient="top", symbolType="stroke", labelFontSize=13)
    .configure_axis(labelFontSize=13)
    .configure_title(fontSize=20)
)

chart
```

```python
chart.save("visuals/chart.png", scale_factor=2)
```
