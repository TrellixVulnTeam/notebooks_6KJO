### Read data file

```python
src = pd.read_csv(
    "data/raw/tabula-July-1-2021-Report-Final.csv", skipfooter=1, engine="python"
)
```

```python
src = src[src["NAME"] != "NAME"]
```

```python
src.columns = src.columns.str.lower().str.replace(" ", "_", regex=False)
```

```python
src["salary"] = (
    src["salary"]
    .str.replace("$", "", regex=False)
    .str.replace(".00", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(int)
)
```

```python
df = src.copy()
```
